```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv(r'C:\Users\admin\Desktop\pie_chart_data.csv')
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10,8))
plt.pie(df.iloc[:,1],
       labels=df.iloc[:,0],
       autopct='%.2f%%',
       textprops={'fontweight':'bold',
                 'size':'x-large',
                 'rotation':'30',
                 'color':'w'})
plt.legend()
plt.title("Cars By Engine Fuel Type",fontsize=18,fontweight='bold')

plt.show()
```
