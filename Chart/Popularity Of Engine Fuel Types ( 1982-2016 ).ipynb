```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv(r'C:\Users\admin\Desktop\stacked_area_chart_data.csv')
```

```python
df.head(2)
```

```python
plt.figure(figsize=(12,8))
sns.set_style('white')
labels=['Diesel','Petrol','Gas']
plt.stackplot(df.iloc[:,0],df.iloc[:,3],df.iloc[:,2],df.iloc[:,1],edgecolor='none')
plt.xticks(df.iloc[:,0],rotation=45)
plt.legend(labels=labels,loc='upper left')
plt.ylabel("Number Of Cars",fontsize=13)
plt.title('Popularity Of Engine Fuel Types ( 1982-2016 )',fontsize=14,fontweight='bold')
sns.despine()
plt.show()
```
