```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv(r'C:\Users\admin\Desktop\scatter_plot_ii (1).csv')
df.head(5)
```

```python
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x='Budget',y='Sales',data=df,scatter_kws={'color':'k'},line_kws={'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title('Effect of Ad Expenditure on Sales',fontsize=14,weight='bold')
plt.show()
```
