```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv(r'C:\Users\admin\Desktop\bar_chart_data.csv')
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df.iloc[:,0],height=df.iloc[:,1],color='midnightblue')
plt.xticks(rotation=45,fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('')
plt.ylabel('Number of Listings',fontsize=13)
plt.title('Cars Listings By Brand',fontsize=16,fontweight='bold')
plt.show()
```
