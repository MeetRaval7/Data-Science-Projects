```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv(r'C:\Users\admin\Desktop\line_chart_data.csv')
```

```python
df['New Date']=pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
labels=['S&p 500','FTSE 100']
plt.figure(figsize=(20,8))
plt.plot(df.iloc[:,3],df.iloc[:,1])
plt.plot(df.iloc[:,3],df.iloc[:,2])
plt.title('S&P vs FTSE Returns (2000-2010)',fontsize=16,fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend(labels=labels,fontsize='large')
plt.show()
```

```python
df1=df[(df['New Date']>='2008-07-01') & (df['New Date']<='2008-12-31') ]
```

```python
labels=['S&p 500','FTSE 100']
plt.figure(figsize=(20,8))
plt.plot(df1.iloc[:,3],df1.iloc[:,1],color='midnightblue')
plt.plot(df1.iloc[:,3],df1.iloc[:,2],color='crimson')
plt.title('S&P vs FTSE Returns (H2 2008)',fontsize=16,fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend(labels=labels,fontsize='large')
plt.show()
```
