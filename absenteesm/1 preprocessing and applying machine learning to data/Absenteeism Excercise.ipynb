# Machine Learning

```python
a=df_preprocessed['Absenteeism Time in Hours'].median()
a
```

```python
targets=np.where(df_preprocessed['Absenteeism Time in Hours']>a,1,0)
```

```python
df_preprocessed['Excessive Absenteeism']=targets
```

```python
df_preprocessed.head()
```

```python
targets.sum()/targets.shape[0]
```

```python
data=df_preprocessed.drop(['Absenteeism Time in Hours','Day Of The Week','Distance to Work','Daily Work Load Average'],axis=1)
```

```python
data.head()
```

```python
unscaled_inputs=data.iloc[:,:-1]
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit(unscaled_inputs)
scaled_inputs=scale.transform(unscaled_inputs)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,random_state=20,test_size=0.2)
```

```python
print(x_train.shape,x_test.shape)
```

# Logistic Regression With Sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
predicted_train=reg.predict(x_train)
```

```python
np.sum((predicted_train==y_train))
```

```python
np.sum((predicted_train==y_train))/predicted_train.shape[0]
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.score(x_test,y_test)
```

```python
predicted_proba=reg.predict_proba(x_test)
```
